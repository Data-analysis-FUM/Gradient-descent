```python
import numpy as np
import random as random
import matplotlib.pyplot as plt
```

```python
def f(x):
    return(x ** 4 + 7 * x ** 3 + 5 * x ** 2 - 17 * x + 3)
```

```python
x = np.arange(-6, 2, 0.01)
y = f(x)
plt.plot(x, y, color = 'red')
plt.grid()
```

```python
def grad_descent(x, gamma = 0.01, iter = 100):
    all_x = [x]
    for i in range(iter):
        x = x - gamma * np.gradient(f(x), x)
        all_x.append(x)
    return(all_x[-1])
```

```python
x = np.random.uniform(-6, 2, 10)
res = grad_descent(x, gamma = 0.01, iter = 100)
```

```python
f(res)
```

```python
x = np.arange(-6, 2, 0.01)
y = f(x)
plt.plot(x, y, color = 'red')
plt.plot(res[-1], f(res[-1]), color = 'blue', marker = 'o') #frist minimum
plt.plot(res[1], f(res[1]), color = 'blue', marker = 'o') #second minimum
plt.grid()
```
